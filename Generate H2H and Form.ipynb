```python
#returns a list with the previous h2h, the most recent at the beginning
def getH2HForMatch(match_id):
    match = matches[matches['id'] == match_id]
    hometeam_id = int(match['home_team_api_id'])
    awayteam_id = int(match['away_team_api_id'])
    match_date = match['date'].values[0]
    #same_matches = matches[((matches['home_team_api_id'] == hometeam_id) & (matches['away_team_api_id'] == awayteam_id))]
    #inverse_matches = matches[((matches['home_team_api_id'] == awayteam_id) & (matches['away_team_api_id'] == hometeam_id))]
    all_matches = matches[(matches['home_team_api_id'] == hometeam_id) & (matches['away_team_api_id'] == awayteam_id) \
                          | ((matches['home_team_api_id'] == awayteam_id) & (matches['away_team_api_id'] == hometeam_id))]
    previous_matches = all_matches[matches['date'] < match_date]
    answer = []
    for index, prev_match in previous_matches.iterrows():
        ht_goal = int(prev_match['home_team_goal'])
        at_goal = int(prev_match['away_team_goal'])
        if ht_goal > at_goal:
            answer.append('H')
        elif ht_goal == at_goal:
            answer.append('D')
        else:
            answer.append('A')
    return list(reversed(answer)) #inversed so that the most recent is in front
```

```python
#returns a dict with the last 5 matches for both teams with the most recent event at the beginning
def getLastFiveGamesForMatch(match_id):
    match = matches[matches['id'] == match_id]
    hometeam_id = int(match['home_team_api_id'])
    awayteam_id = int(match['away_team_api_id'])
    match_date = match['date'].values[0]
    
    return {'home' : getLastFiveGamesForTeam(hometeam_id, match_date), 'away' : getLastFiveGamesForTeam(awayteam_id, match_date)}

def getLastFiveGamesForTeam(team_id, date):
    all_team_matches = matches[(matches['home_team_api_id'] == team_id) | (matches['away_team_api_id'] == team_id)]
    all_team_matches = all_team_matches[all_team_matches['date'] < date]
    results = []
    for match in all_team_matches.values[-5:,:]:
        ht_goal = int(match[9])
        at_goal = int(match[10])
        ht_id = int(match[7])
        at_id = int(match[8])
        #print match
        if ht_goal == at_goal:
            results.append('D')
        elif ((ht_id == team_id) and (ht_goal > at_goal)) or ((at_id == team_id) and (ht_goal < at_goal)):
            results.append('W')
        else:
            results.append('L')
    return list(reversed(results))
```
